```python
# Import library yang dibutuhkan untuk eksplorasi dataset
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

# importing modules
from geopy.geocoders import Nominatim, GoogleV3
```

```python
# Load dataset
df = pd.read_csv('data_california_house.csv')
df.head()
```

```python
df.info()
```

```python
df_loc = df[['longitude','latitude']]
df_loc
```

```python
# # calling the nominatim tool
geoLoc = Nominatim(user_agent="geoapiExercises", timeout=10)

# # passing the coordinates
# # locname = geoLoc.reverse("36.73,-119.79")
# # address = locname.raw['address']
# # city = address.get('city', '')


# print('City : ', city)

road = []
city = []
county = []
state = []

for i in range(len(df_loc["latitude"] + 1)) :
        locname = geoLoc.reverse(f"{df_loc.loc[i]['latitude']}, {df_loc.loc[i]['longitude']}")
        address = locname.raw['address']
        road_raw = address.get('road', '')
        city_raw = address.get('city', '')
        county_raw = address.get('county', '')
        state_raw = address.get('state', '')
        road.append(road_raw)
        city.append(city_raw)
        county.append(county_raw)
        state.append(state_raw)

# printing the address/location name
# print(locname.address)
```

```python
county = pd.DataFrame(county)
county.rename(columns = {0:'county'}, inplace=True)
city = pd.DataFrame(city)
city.rename(columns = {0:'city'}, inplace=True)
road = pd.DataFrame(road)
road.rename(columns = {0:'road'}, inplace=True)
state = pd.DataFrame(state)
state.rename(columns = {0:'state'}, inplace=True)
df_kota = pd.concat([road,county,city,state], axis =1)
df_kota
```

```python
df_kota.to_csv('Data_JALAN_KOTA_DAERAH.csv')
```
